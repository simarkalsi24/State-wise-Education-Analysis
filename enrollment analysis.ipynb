```python
# Find percentage of missing values in each column
null_percent = df.isnull().mean() * 100

# Filter columns with more than 50% null values
high_null_columns = null_percent[null_percent > 50]

print("Columns with more than 50% null values:\n")
print(high_null_columns)
```

```python
# the leasst is with 57% that of followup_status which is around 100000
col_drop = high_null_columns.index.tolist()
col_drop
```

```python
len(col_drop)
```

```python
#droping these columns
df.drop(columns=col_drop, inplace=True)
```

```python
len(df.columns) #10 columns are removed
```

```python
# Find columns with remaining null value
rem_null_columns = df.columns[df.isnull().any()]

# Show data types of those columns
print(df[rem_null_columns].dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check available numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()

# Choose any two numeric columns — update if needed
x_col = numeric_cols[0]
y_col = numeric_cols[1]

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x=x_col, y=y_col, hue='statename', palette='Set1')
plt.title(f'{x_col} vs {y_col} by State')
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.legend(title='State Name')
plt.grid(True)
plt.show()
```

```python
# Select numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Summary of null counts and unique values
summary = pd.DataFrame({
    'null_count': numeric_df.isnull().sum(),
    'unique_count': numeric_df.nunique(),
    'dtype': numeric_df.dtypes
})

# Filter: less than 30% nulls and more than 5 unique values
row_count = len(df)
filtered = summary[
    (summary['null_count'] / row_count < 0.3) & 
    (summary['unique_count'] > 5)
].sort_values(by='unique_count', ascending=False)

filtered
```

```python
import matplotlib.pyplot as plt
df[numeric_cols].hist(bins=20, figsize=(15, 10), grid=False)
plt.suptitle('Distributions of Numeric Columns')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 4))
for i, col in enumerate(['age', 'class', 'socialcategory']):
    plt.subplot(1, 3, i+1)
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='Set2')
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='statename', y='age', palette='Set3')
plt.title('Age Distribution by State')
plt.xlabel('State Name')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```
